作業目標<br>
計算有缺失值的資料，統計量實作<br>
作業重點<br>
當遇到缺失值有函式可以處理，不須額外寫程式刪除<br>
計算統計量時不能出現缺失值

題目:<br>
english_score = np.array([55,89,76,65,48,70])<br>
math_score = np.array([60,85,60,68,np.nan,60])<br>
chinese_score = np.array([65,90,82,72,66,77])<br>
上3列共六位同學的英文、數學、國文成績，第一個元素代表第一位同學，舉例第一位同學英文55分、數學60分、國文65分，今天第五位同學因某原因沒來考試，導致數學成績缺值，運用上列數據回答下列問題。<br>
1. 請計算各科成績平均、最大值、最小值、標準差，其中數學缺一筆資料可忽略?
2. 第五位同學補考數學後成績為55，請計算補考後數學成績平均、最大值、最小值、標準差?
3. 用補考後資料找出與國文成績相關係數最高的學科?

```python
import numpy as np
```

```python
english_score = np.array([55,89,76,65,48,70])
math_score = np.array([60,85,60,68,np.nan,60])
chinese_score = np.array([65,90,82,72,66,77])
```

```python
#1. 請計算各科成績平均、最大值、最小值、標準差，其中數學缺一筆資料可忽略?
print("英文成績平均值：", np.mean(english_score))
print("數學成績平均值：", np.nanmean(math_score))
print("國文成績平均值：", np.mean(chinese_score))
print("英文成績最高分：", np.max(english_score))
print("數學成績最高分：", np.nanmax(math_score))
print("國文成績最高分：", np.max(chinese_score))
print("英文成績最低分：", np.min(english_score))
print("數學成績最低分：", np.nanmin(math_score))
print("國文成績最低分：", np.min(chinese_score))
print("英文成績標準差：", np.std(english_score))
print("數學成績標準差：", np.nanstd(math_score))
print("國文成績標準差：", np.std(chinese_score))
```

```python
#2. 第五位同學補考數學後成績為55，請計算補考後數學成績平均、最大值、最小值、標準差?
math_score[4] = 55
print("數學成績平均值：", np.nanmean(math_score))
print("數學成績最高分：", np.nanmax(math_score))
print("數學成績最低分：", np.nanmin(math_score))
print("數學成績標準差：", np.nanstd(math_score))
```

```python
#3. 用補考後資料找出與國文成績相關係數最高的學科?
chinese_cor_english = np.corrcoef(chinese_score, english_score)
chinese_cor_math = np.corrcoef(chinese_score, math_score)
if chinese_cor_english[0,1] > chinese_cor_math[0, 1]:
    print("英文與國文成績相對係數高")
else:
    print("數學與國文成績相對係數高")
```
